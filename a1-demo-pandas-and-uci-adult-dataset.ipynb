**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year? **

```python
# You code here
print(data[data['salary']=='>50K']['age'].std(),data[data['salary']=='>50K']['age'].mean())
print(data[data['salary']=='<=50K']['age'].std(),data[data['salary']=='<=50K']['age'].mean())
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

```python
# You code here
# data[data['salary']=='>50K']
higher = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters','Doctorate']
data[(data['education'].isin(higher)) & (data['salary']=='>50K')]

print('No it is not compulsory that people having at least higher education earn more than 50K')
```
